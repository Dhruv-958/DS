```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.isna().sum()
```

```python
df["age"].fillna(df["age"].mean())
```

```python
df.drop(columns=["alive"])
```

```python
sns.distplot(x=df["age"], bins=15)
```

```python
sns.jointplot(x=df["age"], y=df["fare"], kind='scatter')
```

```python
sns.jointplot(x=df["age"], y=df["fare"], kind='hex')
```

```python
sns.rugplot(x=df["fare"])
```

```python
sns.barplot(x=df["age"], y=df["sex"])
```

```python
sns.boxplot(x=df["age"], y=df["sex"])
```

```python
sns.countplot(x=df["sex"])
```

```python
sns.violinplot(x=df["age"],y=df["sex"])
```

```python
sns.stripplot(x=df["age"], y=df["sex"], jitter=True)
```

```python
sns.swarmplot(x=df["age"], y=df["sex"],hue=df["survived"])
```
