# **Z_SCORES**

```python
data1 = np.array([2,4,5,6,3,5,67,3,650,780,100])
z_scores = stats.zscore(data1)
t = 1.5
outliers = np.where(np.abs(z_scores) > t)[0]
z_scores
```

```python
print("Outliers: ", data1[outliers])
```

```python
plt.boxplot(data1)
```

# **IQR**

```python
data2 = np.array([2,4,5,6,3,5,14,3,17,20,100])
data2
```

```python
plt.boxplot(data2)
```

```python
new_data2 = data2[data2 < 100]
new_data2
```

```python
plt.boxplot(new_data2)
```

```python
median = np.median(data2)
median
```

```python
Q1 = [i for i in data2 if i<median]
Q3 = [i for i in data2 if i>median]
print("Q1: ", Q1, "\nQ3: ", Q3)
```

```python
Q1_med = np.median(Q1)
Q3_med = np.median(Q3)
print("Q1: ", Q1_med, "\nQ3: ", Q3_med)
IQR = Q3_med - Q1_med
print("IQR: ", IQR)
```

```python
new_data2 = data2[data2!=100]
print(new_data2)
plt.boxplot(new_data2)
```

# **Normalizing Values**

```python
data3 = data2
print(data3)
min = np.min(data3)
max = np.max(data3)
print("Min: ", min, "\nMax: ", max)
```

```python
normalized_values = [(i-min)/(max-min) for i in data3]
print(normalized_values)
```
