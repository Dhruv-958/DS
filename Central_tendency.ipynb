```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_Data.csv")
df
```

```python
df.describe()
```

```python
df['Salary'].mean()
```

```python
df1 = df.groupby(['Salary','Age'])
```

```python
df1.describe()
```

```python
df.isna().sum()
```

```python
df1 = df.dropna()
df1.isna().sum()
```

```python
df1
```

```python
def age_group(age) :
  if age <= 39:
    return "Young Adult"
  elif age <=59:
    return "Middle Age"
  else:
    return "Old Adult"

df1["Age"] = df1["Age"].apply(age_group)
df1
```

```python
df1["Age"].value_counts()
```

```python
df1.groupby(["Age"])["Salary"].mean()
```

```python
df1.groupby(["Age"])["Salary"].agg(pd.Series.mode)
```

```python
df1.groupby(["Age"])["Salary"].median()
```

```python
min = df1.groupby(["Age"])["Salary"].min()
min
```

```python
max = df1.groupby(["Age"])["Salary"].max()
max
```

```python
range = max - min
range
```

```python
std_dev = df1.groupby(["Age"])["Salary"].std()
std_dev
```

```python
var = np.square(std_dev)
var
```
