```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score, precision_score, recall_score
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.isna().sum()
```

```python
x = df.drop(columns=['species'])
y = df['species']
print(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state = 32)
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
model.score(x_test,y_predict)
```

```python
cnf = confusion_matrix(y_test,y_predict)
cnf
```

```python
print("Accuracy: ", accuracy_score(y_test,y_predict),"\nRecall: ", recall_score(y_test, y_predict, average='weighted'), "\nPrecision: ",precision_score(y_test,y_predict, average='weighted'))
```
